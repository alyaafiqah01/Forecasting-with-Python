```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("malaysia_parcel_volumes_full.csv",delimiter=",")
```

```python
data.describe()
```

```python
data.set_index('Date', inplace=True)
```

```python
X = data.columns.drop('Transportation Costs')
```

```python
X = np.array(data[X])
```

```python
y = np.array(data['Transportation Costs'])
```

```python
model = LinearRegression().fit(X,y)
```

```python
coef = model.coef_
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=data.columns.drop('Transportation Costs'), y=coef)
plt.xlabel("Features")
plt.ylabel("Coefficients")
plt.title("Coefficients of Linear Regression Model")
plt.show()
```

```python
for feature in data.columns.drop('Transportation Costs'):
    plt.figure(figsize=(10,6))
    plt.scatter(data[feature], data['Transportation Costs'])
    plt.xlabel(feature)
    plt.ylabel('Transportation Costs')
    plt.title('{} vs Transportation Costs'.format(feature))
    plt.show()
```
